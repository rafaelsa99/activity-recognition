<b>Activity Recognition Using Wearable Physiological Measurements</b>: https://www.mdpi.com/1424-8220/19/24/5524<br>
<b>Dataset</b>: http://www.mdpi.com/1424-8220/19/24/5524/s1 <br><br>
O problema consiste em analisar um conjunto de <i>features</i> de forma a extrair informação a partir de:
- <i>electrocardiogram</i> (<b>ECG</b>);
- <i>thoracic electrical bioimpedance</i> (<b>TEB</b>);
- <i>electrodermal activity signals</i> (<b>EDA</b>). <br>

Os dados foram extraídos de 40 pessoas com base em estímulos dados por filmes, atividades diárias, jogos, atividade física, entre outros. Para extrair os dados relativos ao ECG e TEB foi utilizada uma luva e um colete e para a EDA uma bracelete. No total foram extraídas 533 <i>features</i>. <br>
Com base nos dados obtidos deve ser reconhecido o tipo de atividade que estava a ser exercida. As atividades analisadas poderão ser:
- <i>neutral</i>;
- <i>emotional</i>;
- <i>mental</i>;
- <i>physical</i>.

O reconhecimento da atividade poderá ser aplicado na prevenção de, por exemplo, erros humanos em profissões de alto risco.

```python
import numpy as np
import pandas as pd
%matplotlib inline 
import matplotlib.pyplot as plt 
import matplotlib.gridspec as gridspec
import itertools
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn import svm, datasets
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, plot_confusion_matrix
from sklearn import neighbors

import warnings
warnings.filterwarnings("ignore")
```

```python
combinations = ["ECG_TEB_EDA", "ECG", "TEB", "EDA", "ECG_TEB", "ECG_EDA", "TEB_EDA"]
df = pd.read_csv("data_modified.csv")
df
```

#### Map dos tipos de atividades
Nos dados fornecidos, a coluna correspondente à classificação que se pretende obter vem em formato numérico. Considera-se que, neste contexto, não existe a noção de ordem nas atividades e portanto assume-se que não faz sentido esta classificação estar no formato numérico. Posto isto, os valores da coluna são alterados para as correspondentes labels.

```python
activity_mapping = {
 1: 'neutral',
 2: 'emotional',
 3: 'mental',
 4: 'physical'
}
df['Activity'] = df['Activity'].map(activity_mapping)
df
```

#### Separação entre categorias de features (ECG-TEB-EDA)
Para depois se proceder à comparação entre categorias de features, é necessário separar as combinações de categorias em vários dataframes.

```python
df_ECG = df.filter(regex=("ECG_.*|Activity"))
df_TEB = df.filter(regex=("IT_.*|Activity"))
df_EDA = df.filter(regex=("EDA_.*|Activity"))
df_ECG_TEB = df.filter(regex=("(ECG_.*)|(IT_.*)|(Activity)"))
df_ECG_EDA = df.filter(regex=("(ECG_.*)|(EDA_.*)|(Activity)"))
df_TEB_EDA = df.filter(regex=("(IT_.*)|(EDA_.*)|(Activity)"))
```

#### Separar a classificação do resto da informação de cada instância
Nesta secção separa-se as colunas de features da coluna do tipo da atividade, pois para o treino e classificação é necessário estarem separados.

```python
target_results = df[['Activity']]
target_results_ECG = df_ECG[['Activity']]
target_results_TEB = df_TEB[['Activity']]
target_results_EDA = df_EDA[['Activity']]
target_results_ECG_TEB = df_ECG_TEB[['Activity']]
target_results_ECG_EDA = df_ECG_EDA[['Activity']]
target_results_TEB_EDA = df_TEB_EDA[['Activity']]
target_results
```

No bloco anterior, a coluna do tipo de atividade é copiada para uma variável à parte. Por último é necessário remover a coluna do dataframe inicial.

```python
df = df.drop("Activity", axis=1)
df_ECG = df_ECG.drop("Activity", axis=1)
df_TEB = df_TEB.drop("Activity", axis=1)
df_EDA = df_EDA.drop("Activity", axis=1)
df_ECG_TEB = df_ECG_TEB.drop("Activity", axis=1)
df_ECG_EDA = df_ECG_EDA.drop("Activity", axis=1)
df_TEB_EDA = df_TEB_EDA.drop("Activity", axis=1)
df
```

#### Separação em dados de treino e teste
Para validar a classificação depois do treino é utilizado o método de cross-validation e, para isso, são necessários alguns dados de teste. Portanto, divide-se a totalidade dos dados em 70% para treino e 30% para teste.

```python
data_train, data_test, target_train, target_test  = train_test_split( df, target_results, test_size=0.3, random_state=0)
data_train_ECG, data_test_ECG, target_train_ECG, target_test_ECG = train_test_split( df_ECG, target_results_ECG, test_size=0.3, random_state=0)
data_train_TEB, data_test_TEB, target_train_TEB, target_test_TEB = train_test_split( df_TEB, target_results_TEB, test_size=0.3, random_state=0)
data_train_EDA, data_test_EDA, target_train_EDA, target_test_EDA = train_test_split( df_EDA, target_results_EDA, test_size=0.3, random_state=0)
data_train_ECG_TEB, data_test_ECG_TEB, target_train_ECG_TEB, target_test_ECG_TEB = train_test_split( df_ECG_TEB, target_results_ECG_TEB, test_size=0.3, random_state=0)
data_train_ECG_EDA, data_test_ECG_EDA, target_train_ECG_EDA, target_test_ECG_EDA = train_test_split( df_ECG_EDA, target_results_ECG_EDA, test_size=0.3, random_state=0)
data_train_TEB_EDA, data_test_TEB_EDA, target_train_TEB_EDA, target_test_TEB_EDA = train_test_split( df_TEB_EDA, target_results_TEB_EDA, test_size=0.3, random_state=0)
```

#### Normalização
Depois de se dividir os dados é necessário normalizá-los pois, para alguns classificadores, este é um passo importante para uma boa precisão do classificador. Neste caso, foram utilizados os classificadores Random Forest e KNN. Para o Random Forest este passo não é muito importante, uma vez que não tem um efeito significativo na precisão. Contudo, o KNN foi utilizado e, portanto, os dados foram centrados e normalizados.

```python
sc = StandardScaler()
data_combinations = {}

data_train_normalized = sc.fit_transform(data_train)
data_test_normalized = sc.transform(data_test)
data_train_ECG_normalized = sc.fit_transform(data_train_ECG)
data_test_ECG_normalized = sc.transform(data_test_ECG)
data_train_TEB_normalized = sc.fit_transform(data_train_TEB)
data_test_TEB_normalized = sc.transform(data_test_TEB)
data_train_EDA_normalized = sc.fit_transform(data_train_EDA)
data_test_EDA_normalized = sc.transform(data_test_EDA)
data_train_ECG_TEB_normalized = sc.fit_transform(data_train_ECG_TEB)
data_test_ECG_TEB_normalized = sc.transform(data_test_ECG_TEB)
data_train_ECG_EDA_normalized = sc.fit_transform(data_train_ECG_EDA)
data_test_ECG_EDA_normalized = sc.transform(data_test_ECG_EDA)
data_train_TEB_EDA_normalized = sc.fit_transform(data_train_TEB_EDA)
data_test_TEB_EDA_normalized = sc.transform(data_test_TEB_EDA)

data_combinations[combinations[0]] = {"train": {"data": data_train_normalized, "target": target_train, "unnormalized": data_train}, "test": {"data": data_test_normalized, "target": target_test, "unnormalized": data_test}}
data_combinations[combinations[1]] = {"train": {"data": data_train_ECG_normalized, "target": target_train_ECG, "unnormalized": data_train_ECG}, "test": {"data": data_test_ECG_normalized, "target": target_test_ECG, "unnormalized": data_test_ECG}}
data_combinations[combinations[2]] = {"train": {"data": data_train_TEB_normalized, "target": target_train_TEB, "unnormalized": data_train_TEB}, "test": {"data": data_test_TEB_normalized, "target": target_test_TEB, "unnormalized": data_test_TEB}}
data_combinations[combinations[3]] = {"train": {"data": data_train_EDA_normalized, "target": target_train_EDA, "unnormalized": data_train_EDA}, "test": {"data": data_test_EDA_normalized, "target": target_test_EDA, "unnormalized": data_test_EDA}}
data_combinations[combinations[4]] = {"train": {"data": data_train_ECG_TEB_normalized, "target": target_train_ECG_TEB, "unnormalized": data_train_ECG_TEB}, "test": {"data": data_test_ECG_TEB_normalized, "target": target_test_ECG_TEB, "unnormalized": data_test_ECG_TEB}}
data_combinations[combinations[5]] = {"train": {"data": data_train_ECG_EDA_normalized, "target": target_train_ECG_EDA, "unnormalized": data_train_ECG_EDA}, "test": {"data": data_test_ECG_EDA_normalized, "target": target_test_ECG_EDA, "unnormalized": data_test_ECG_EDA}}
data_combinations[combinations[6]] = {"train": {"data": data_train_TEB_EDA_normalized, "target": target_train_TEB_EDA, "unnormalized": data_train_TEB_EDA}, "test": {"data": data_test_TEB_EDA_normalized, "target": target_test_TEB_EDA, "unnormalized": data_test_TEB_EDA}}

data_train
```

O processo de seleção de features foi feito utilizando o Random Forest. Após o treino do classificador é recolhida a importância que o Random Forest atribuiu de todas as features e, apartir daí, é feita a escolha das mais importantes.

A primeira seleção de features que foi feita utilizou todas as três categorias de features. Depois desta seleção, escolheu-se também as melhores para combinações de apenas duas ou uma categoria. ## Classificadores

Foram testados dois classificadores:

### Configuração dos Classificadores
#### KNN: K-Nearest Neighbor
Para definir o número de vizinhos (parâmetro k) para se utilizar nos testes com o KNN, testou-se o erro médio de classificação consoante a variação do parâmetro k para as várias combinações de categorias de <i>features</i>. O parâmetro k varia entre 1 e 10 vizinhos, e foi utilizada a <i>weight function distance</i>. <br>
Os testes foram executados com as 10 melhores <i>features</i> de cada combinação, uma vez que a maioria dos testes que serão realizados com o classificador terão 10 ou mais <i>features</i>.<br>


```python

best_features = [feat_importances, feat_importances_ECG, feat_importances_TEB, feat_importances_EDA, feat_importances_ECG_TEB, feat_importances_ECG_EDA, feat_importances_TEB_EDA]
ECG_TEB_EDA = [data_train, target_train, data_test, target_test]
ECG = [data_train_ECG, target_train_ECG, data_test_ECG, target_test_ECG]
TEB = [data_train_TEB, target_train_TEB, data_test_TEB, target_test_TEB]
EDA = [data_train_EDA, target_train_EDA, data_test_EDA, target_test_EDA]
ECG_TEB = [data_train_ECG_TEB, target_train_ECG_TEB, data_test_ECG_TEB, target_test_ECG_TEB]
ECG_EDA = [data_train_ECG_EDA, target_train_ECG_EDA, data_test_ECG_EDA, target_test_ECG_EDA]
TEB_EDA = [data_train_TEB_EDA, target_train_TEB_EDA, data_test_TEB_EDA, target_test_TEB_EDA]

labels = ["ECG+ETB+EDA", "ECG", "TEB", "EDA", "ECG+TEB", "ECG+EDA", "TEB+EDA"]

gs = gridspec.GridSpec(4, 2)

fig = plt.figure(figsize=(14, 12))

for dataknn, data, lab, feat, grd in zip([[df, target_results], [df_ECG, target_results_ECG], [df_TEB, target_results_TEB], [df_EDA, target_results_EDA], [df_ECG_TEB, target_results_ECG_TEB], [df_ECG_EDA, target_results_ECG_EDA], [df_TEB_EDA, target_results_TEB_EDA]],[ECG_TEB_EDA, ECG, TEB, EDA, ECG_TEB, ECG_EDA, TEB_EDA], labels, best_features, [[0,0], [0,1], [1,0], [1,1],[2,0],[2,1],[3,0]]):
    error = []
    temp = []
    for feature in feat.nlargest(10).iteritems():
        temp.append(feature[0])
    df_selected = dataknn[0].filter(temp, axis=1)
    d_train, d_test, t_train, t_test = train_test_split( df_selected, dataknn[1], test_size=0.3, random_state=0)
    d_train = sc.fit_transform(d_train)
    d_test = sc.transform(d_test)
    # Calculating error for K values between 1 and 10
    for i in range(1, 10):
        knn = neighbors.KNeighborsClassifier(n_neighbors=i, weights='distance')
        knn.fit(d_train, np.ravel(t_train))
        pred_i = knn.predict(d_test)
        error.append(np.mean(pred_i != np.ravel(t_test)))

    ax = plt.subplot(gs[grd[0], grd[1]])
    fig = plt.plot(range(1, 10), error, color='red', linestyle='dashed', marker='o',
            markerfacecolor='blue', markersize=10, label='KNN')
    plt.title('Error Rate K Value using ' + lab + " feature set")
    plt.legend()
    plt.xlabel('K Value')
    plt.ylabel('Mean Error')
    plt.grid()
    plt.tight_layout()
    
plt.show()
```

Através de uma análise aos gráficos, é possível concluir que, de uma forma geral, quanto maior o número de vizinhos, maior será o erro médio de classificação, sendo que os melhores resultados foram obtidos para k = 1 e k = 2. Os valores de k são baixos, uma vez que vetores de <i>features</i> não são ideais neste classificador. 
Posto isto, definiu-se o valor do parâmetro k como 1 vizinho pois, apesar de não ser um valor comum neste classificador, foi o que apresentou o menor erro de classificação. Não foi escolhido o valor de 2 vizinhos pois poderia originar empates na decisão de classificação.

#### Random Forest
Para efetuar os vários testes utilizando número de <i>features</i> diferentes foi utilizado o parâmetro 'max_features' com o valor pretendido. 

### Comparação de Classificadores

Para comparar o desempenho dos dois classificadores, foram efetuados testes para as várias combinações de categorias de <i>features</i>. Cada combinação foi testada com diferentes números de <i>features</i>, e foi registado o respetivo <i>error rate</i>, em percentagem.



```python

best_features = [feat_importances, feat_importances_ECG, feat_importances_TEB, feat_importances_EDA, feat_importances_ECG_TEB, feat_importances_ECG_EDA, feat_importances_TEB_EDA]

gs = gridspec.GridSpec(4, 2)

fig = plt.figure(figsize=(17, 12))

for dataknn, data, lab, feat, grd in zip([[df, target_results], [df_ECG, target_results_ECG], [df_TEB, target_results_TEB], [df_EDA, target_results_EDA], [df_ECG_TEB, target_results_ECG_TEB], [df_ECG_EDA, target_results_ECG_EDA], [df_TEB_EDA, target_results_TEB_EDA]],[ECG_TEB_EDA, ECG, TEB, EDA, ECG_TEB, ECG_EDA, TEB_EDA], labels, best_features, [[0,0], [0,1], [1,0], [1,1],[2,0],[2,1],[3,0]]):
    error = []
    for i in range(5, 150, 10):
        temp = []
        for feature in feat.nlargest(i).iteritems():
            temp.append(feature[0])
        df_selected = dataknn[0].filter(temp, axis=1)
        d_train, d_test, t_train, t_test = train_test_split( df_selected, dataknn[1], test_size=0.3, random_state=0)
        d_train = sc.fit_transform(d_train)
        d_test = sc.transform(d_test)
        knn = neighbors.KNeighborsClassifier(n_neighbors=1,weights='distance', algorithm='auto')
        knn.fit(d_train, np.ravel(t_train))
        score = knn.score(d_test, np.ravel(t_test))
        error_rate = (1 - score) * 100
        error.append(error_rate)
        

    ax = plt.subplot(gs[grd[0], grd[1]])
    fig = plt.plot(range(5, 150, 10), error, color='red', linestyle='dashed', marker='o',
                   markerfacecolor='blue', markersize=10, label='KNN')
    
    error = []
    for i in range(5, 150, 10):
        rf = RandomForestClassifier(max_features=i, oob_score=True)
        rf.fit(data[0], np.ravel(data[1]))
        score = rf.score(data[2], np.ravel(data[3]))
        error_rate = (1 - score) * 100
        error.append(error_rate)
    
    fig = plt.plot(range(5, 150, 10), error, color='yellow', linestyle='dashed', marker='*',
                   markerfacecolor='green', markersize=10, label='Random Forest')
    plt.title('Classifiers comparison using ' + lab + ' feature set')
    plt.xlabel('Number of selected features')
    plt.ylabel('Error rate (%)')
    plt.grid()
    plt.legend()
    plt.tight_layout()
    
plt.show()
```

Os resultados obtidos demonstram que o Random Forest é o classificador que obtém os melhores resultados em todas as combinações de categorias de <i>features</i>. <br>
